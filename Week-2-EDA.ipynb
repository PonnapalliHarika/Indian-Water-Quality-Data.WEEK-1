## 4.Visualizations

## a.Correlation Heatmap

```python
# Select only numeric columns
numeric_df = df.select_dtypes(include=['float64','int64'])

# Compute correlation
corr = numeric_df.corr()

# Plot heatmap
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(15,10))
sns.heatmap(corr, cmap="coolwarm", annot=False, linewidths=0.5)
plt.title("Correlation Heatmap of Water Quality Parameters", fontsize=16)
plt.show()
```

## b.Boxplots of Key Parameters

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=df[["pH - Max","BOD (mg/L) - Max","Dissolved - Max"]])
plt.title("Boxplots of Key Water Quality Parameters")
plt.show()
```

## c. Pairplot

```python
sns.pairplot(df[["pH - Max","BOD (mg/L) - Max","Dissolved - Max"]])
plt.show()
```

```python
print("Sample Insights:")
print("- pH values are mostly within normal range, but some outliers exist.")
print("- BOD (pollution indicator) shows variation, with some high values suggesting pollution in certain sites.")
print("- Dissolved Oxygen is negatively correlated with BOD, as expected (more pollution = less oxygen).")
```
