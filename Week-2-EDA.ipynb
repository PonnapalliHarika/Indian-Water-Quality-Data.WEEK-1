## 4.Visualizations

## a.Correlation Heatmap

```python
# Select only numeric columns
numeric_df = df.select_dtypes(include=['float64','int64'])

# Compute correlation
corr = numeric_df.corr()

# Plot heatmap
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(15,10))
sns.heatmap(corr, cmap="coolwarm", annot=False, linewidths=0.5)
plt.title("Correlation Heatmap of Water Quality Parameters", fontsize=16)
plt.show()
```

## b.Boxplots of Key Parameters

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=df[["pH - Max","BOD (mg/L) - Max","Dissolved - Max"]])
plt.title("Boxplots of Key Water Quality Parameters")
plt.show()
```

## c. Pairplot

```python
sns.pairplot(df[["pH - Max","BOD (mg/L) - Max","Dissolved - Max"]])
plt.show()
```

## d. Histograms for all numeric columns

```python
df.hist(bins=20, figsize=(18, 15))
plt.suptitle("Distribution of Water Quality Parameters", fontsize=16)
plt.tight_layout()
plt.show()
```

## e. State-wise Comparison (example: BOD levels)

```python
plt.figure(figsize=(15, 6))
sns.barplot(x="State Name", y="pH - Max", data=df, estimator=np.mean, errorbar=None)
plt.xticks(rotation=90)
plt.title("Average Maximum pH by State", fontsize=14)
plt.show()
```

## f. State-wise Comparison (example: Nitrate levels)

```python
plt.figure(figsize=(15, 6))
sns.barplot(x="State Name", y="NitrateN (mg/L) - Max", data=df, estimator=np.mean, errorbar=None)
plt.xticks(rotation=90)
plt.title("Average Maximum Nitrate by State", fontsize=14)
plt.show()
```
